# Correlation Analysis
<p>The purpose of this analysis is to find the correlation between various cryptocoins.The crypto data were all gotten from <code>Yahoo finance</code> using the <code>Pandas Data Reader</code> for its extraction. <code>Seaborn</code> was used to visualise the correlation.</p>
<p>The Cryptocurrencies that would be used in this analysis are Bitcoin <code>BTC</code>, Ethereum <code>ETH</code>, Binance Coin, <code>BNB</code>, Ripple <code>XRP</code>, Cardano <code>ADA</code>, Dogecoin <code>DOGE</code>, Bitcoin Cash <code>BCH</code>, Litecoin <code>LTC</code>, Youcash <code>YOUC</code> and USD Coin <code>USDC</code>.</p> 

```python
# importing relevant libraries

import numpy as np
import pandas as pd
import pandas_datareader as pdd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sb

# to ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Choosing dates to extract data
start = dt.datetime(2017, 1, 1)
end = dt.datetime.now()
```

```python
# Extracting BTC 

btc = pdd.DataReader('BTC-USD', 'yahoo', start, end)
btc.reset_index(inplace=True)
crypto = btc[['Date', 'Close']]
crypto = crypto.rename(columns={'Close': 'BTC'})
```

```python
# Extracting ETH

eth = pdd.DataReader('ETH-USD', 'yahoo', start, end)
eth.reset_index(inplace=True)
crypto['ETH'] = eth['Close']
```

```python
# Extracting BNB 

bnb = pdd.DataReader('BNB-USD', 'yahoo', start, end)
bnb.reset_index(inplace=True)
crypto['BNB'] = bnb['Close']
```

```python
# Extracting XRP

xrp = pdd.DataReader('XRP-USD', 'yahoo', start, end)
xrp.reset_index(inplace=True)
crypto['XRP'] = xrp['Close']
```

```python
# Extracting Cardano

ada = pdd.DataReader('ADA-USD', 'yahoo', start, end)
ada.reset_index(inplace=True)
crypto['ADA'] = ada['Close']
```

```python
# Extracting Dogecoin

doge = pdd.DataReader('DOGE-USD', 'yahoo', start, end)
doge.reset_index(inplace=True)
crypto['DOGE'] = doge['Close']
```

```python
# Extracting Bitcoin Cash

bch = pdd.DataReader('BCH-USD', 'yahoo', start, end)
bch.reset_index(inplace=True)
crypto['BCH'] = bch['Close']
```

```python
# Extracting Litecoin

ltc = pdd.DataReader('LTC-USD', 'yahoo', start, end)
ltc.reset_index(inplace=True)
crypto['LTC'] = ltc['Close']
```

```python
# Extracting Youcash

youc = pdd.DataReader('YOUC-USD', 'yahoo', start, end)
youc.reset_index(inplace=True)
crypto['YOUC'] = youc['Close']
```

```python
# Extracting USDC

usdc = pdd.DataReader('USDC-USD', 'yahoo', start, end)
usdc.reset_index(inplace=True)
crypto['USDC'] = usdc['Close']
```

```python
# to show the first 5 values of the merged data
crypto.head()
```

```python
# Makes date the index
crypto.set_index('Date', inplace=True)
crypto.head()
```

```python
crypto.columns
```

```python
# To show the correlation

corrs = crypto.corr()
corrs
```

```python
# To plot the correlation

plt.figure(figsize=(20,10))
sb.heatmap(corrs, annot=True)
plt.title('Heatmap showing the correlation between different Cryptocurrency coins', fontsize=15)
plt.show()
```

<p> The results from this correlation analysis shows that the cryptocoins chosen have some form of relationship between them.The heatmap above helps to clearly visualize how correlated they are. The major coins that would compared against are <code>BTC</code> and <code>ETH</code>.</p>
<p> From the Heatmap, Litecoin <code>LTC</code> has the highest correlation with Bitcoin <code>BTC</code> with a correlation value of <em>0.73</em> followed by Binance <code>BNB</code>, <em>0.66</em> and Ethereum <code>ETH</code>, <em>0.62</em>. Also, Bitcoin cash <code>BCH</code> and USD coin <code>USDC</code> with correlation values of <em>-0.29</em> and <em>-0.35</em> respectively are negatively correlated with Bitcoin.</p>

<p> The cryptocoins that have the highest correlation with Ethereum <code>ETH</code> are Binance <code>BNB</code>: <em>0.96</em>, Cardano <code>ADA</code>: <em>0.92</em> and Dogecoin <code>DOGE</code>: <em>0.86</em>. While Youcash <code>YOUC</code> and USD coin <code>USDC</code> are negatively correlated with Ethereum with the correlation values of <em> -0.52</em> and <em>-0.081</em> respectively.
